```python
df = pd.read_csv("./data/scMultiSim_final/scmultisim_dataframe.csv")
neighbors = pd.read_csv("./data/scMultiSim_final/simulated/neighbors.csv")
gt = pd.read_csv("./data/scMultiSim_final/simulated/cci_gt.csv").iloc[:,-2:]
combined_ctgt = gt["ct1"].astype(str) + "|" + gt["ct2"].astype(str)
gt_set = set(combined_ctgt)

scmultisim_adj = np.zeros(shape=(len(df),len(df)))

edge_list = []

edge_x = []
edge_y = []

for i in range(len(df)):
    cell1_row = df.iloc[i]
    cell1_neighbors = list(neighbors.iloc[i][neighbors.iloc[i].notnull()].astype(int))
    cell1_type = cell1_row["Cell_Type"].split("_")[-1]
    
    for j in cell1_neighbors:
        cell2_idx = int(j)-1
        cell2_row = df.iloc[cell2_idx]
        cell2_type =  cell2_row["Cell_Type"].split("_")[-1]
        
        # combined_ct= f"{cell1_type}|{cell2_type}"
        # # print(cell2_idx,combined_ct)
        
        # if combined_ct in gt_set:
        scmultisim_adj[i][cell2_idx] = 1
        scmultisim_adj[cell2_idx][i] = 1
        edge_list.append([int(i), int(cell2_idx)])
        edge_list.append([int(cell2_idx),int(i)])
            
        
edge_arr = np.array(edge_list).T
```

```python
edge_arr
```

```python
G = nx.from_edgelist(edge_arr.T)
G.add_nodes_from(range(1200))

node_positions = df[["X","Y"]]

for i, rowloc in node_positions.iterrows():
    G.nodes[i]['pos'] = (rowloc["X"],rowloc["Y"])
    
edge_x = []
edge_y = []
for edge in G.edges():
    x0, y0 = G.nodes[edge[0]]['pos']
    x1, y1 = G.nodes[edge[1]]['pos']
    edge_x.append(x0)
    edge_x.append(x1)
    edge_x.append(None)
    edge_y.append(y0)
    edge_y.append(y1)
    edge_y.append(None)

```

```python
edge_arr = nx.to_pandas_edgelist(G).values.T
adjacency = nx.to_pandas_adjacency(G)

# adjacency.to_csv("./data/scMultiSim_final/adj.csv",index=False)
edge_arr
```
